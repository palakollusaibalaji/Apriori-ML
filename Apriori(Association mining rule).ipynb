```python
def encode(x):
    if isinstance(x,str): #check if x is a string
         try:
             x=float(x) # Attempt to convert to number
         except ValueError:
             return 0 #if conversion to number fails,return 0
    if x<=0: 
        return 0 ; 
    if x>=1: 
        return 1; 

basket = basket_df.applymap(encode)
basket.head()
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Assuming basket is your datafrome 
frequent_itemsets = apriori(basket,min_support=0.07,use_colnames=True)
frequent_itemsets
```

```python
#convert fronzenset to list for itemsets column 
frequent_itemsets["itemsets"] = frequent_itemsets["itemsets"].apply(list)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets,metric="lift",min_threshold=0.5,num_itemsets=frequent_itemsets.shape[0])
rules.head(20)
```

```python
rules[(rules["lift"] >= 1) & (rules["confidence"]>=0.05)].head()
```
